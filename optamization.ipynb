```python
import numpy as np
import matplotlib.pyplot as plt

#(a) Python code for generating a line plot of the function:

def f(λ):
    return np.sin(λ)

λ = np.linspace(0, 4*np.pi, 100)
y = f(λ)

plt.plot(λ, y)
plt.xlabel('λ')
plt.ylabel('f(λ)')
plt.title('Plot of f(λ) = sin(λ)')
plt.show()

#(b) Loop to evaluate 20 iterations with λ(0) = 0.2 and η = 0.1:
λ = 0.2
η = 0.1
iterations = 20
update_values = []

for i in range(iterations):
    update = η * np.cos(λ)  # The derivative of sin(λ) is cos(λ)
    λ = λ + update
    update_values.append(λ)

plt.plot(range(iterations), update_values)
plt.xlabel('Iteration')
plt.ylabel('λ')
plt.title('Update Values for λ(0) = 0.2 and η = 0.1')
plt.show()

#(c) Comparing the update values for different learning rates:
λ = 0.2
η_values = [0.1, 0.75, 0.2]
iterations = 20
update_values = []

for η in η_values:
    λ = 0.2
    temp_update_values = []
    for i in range(iterations):
        update = η * np.cos(λ)
        λ = λ + update
        temp_update_values.append(λ)
    update_values.append(temp_update_values)

plt.figure(figsize=(10, 5))
for i, η in enumerate(η_values):
    plt.plot(range(iterations), update_values[i], label='η = ' + str(η))
plt.xlabel('Iteration')
plt.ylabel('λ')
plt.title('Update Values for λ(0) = 0.2')
plt.legend()
plt.show()

#(d) Performing the same steps for λ(0) = 3.5:
λ = 3.5
η_values = [0.1, 0.75, 0.2]
iterations = 20
update_values = []

for η in η_values:
    λ = 3.5
    temp_update_values = []
    for i in range(iterations):
        update = η * np.cos(λ)
        λ = λ + update
        temp_update_values.append(λ)
    update_values.append(temp_update_values)

plt.figure(figsize=(10, 5))
for i, η in enumerate(η_values):
    plt.plot(range(iterations), update_values[i], label='η = ' + str(η))
plt.xlabel('Iteration')
plt.ylabel('λ')
plt.title('Update Values for λ(0) = 3.5')
plt.legend()
plt.show()


```

```python
#(a) To plot the function g(x) = x^3 - 2x^2 + 1 over the interval x ∈ [-2, 3]:
import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x)
def g(x):
    return x**3 - 2*x**2 + 1

# Generate x values in the interval [-2, 3]
x = np.linspace(-2, 3, 100)

# Compute the corresponding y values using g(x)
y = g(x)

# Plot the function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Plot of g(x)')
plt.grid(True)
plt.show()

#(b) Implementing the gradient ascent algorithm with a learning rate of η = 0.01 and 
#starting value x(0) = -1, and computing 10 iterations to plot the update values against 
#the iteration number:

import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x) and its derivative
def g(x):
    return x**3 - 2*x**2 + 1

def g_prime(x):
    return 3*x**2 - 4*x

# Gradient Ascent Algorithm
def gradient_ascent(eta, x0, num_iterations):
    x_values = [x0]
    for i in range(num_iterations):
        x = x_values[-1]
        x_new = x + eta * g_prime(x)
        x_values.append(x_new)
    return x_values

# Set the learning rate, starting value, and number of iterations
eta = 0.01
x0 = -1
num_iterations = 10

# Perform gradient ascent and obtain the update values
x_values = gradient_ascent(eta, x0, num_iterations)

# Plot the update values against the iteration number
iterations = range(num_iterations + 1)
plt.plot(iterations, x_values, marker='o')
plt.xlabel('Iteration')
plt.ylabel('x')
plt.title('Gradient Ascent')
plt.grid(True)
plt.show()


#(c) Repeating part (b) with a learning rate of η = 0.1 and starting value x(0) = 1 to observe 
#the convergence behavior:

import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x) and its derivative
def g(x):
    return x**3 - 2*x**2 + 1

def g_prime(x):
    return 3*x**2 - 4*x

# Gradient Ascent Algorithm
def gradient_ascent(eta, x0, num_iterations):
    x_values = [x0]
    for i in range(num_iterations):
        x = x_values[-1]
        x_new = x + eta * g_prime(x)
        x_values.append(x_new)
    return x_values

# Set the learning rate, starting value, and number of iterations
eta = 0.1
x0 = 1
num_iterations = 10

# Perform gradient ascent and obtain the update values
x_values = gradient_ascent(eta, x0, num_iterations)

# Plot the update values against the iteration number
iterations = range(num_iterations + 1)
plt.plot(iterations, x_values, marker='o')
plt.xlabel('Iteration')
plt.ylabel('x')
plt.title('Gradient Ascent')
plt.grid(True)
plt.show()


```

```python
#(a) To generate a line plot of the function h(t) = e^(-t^2/2) / √(2π) over the interval t ∈ [-3, 3]:

import numpy as np
import matplotlib.pyplot as plt

# Define the function h(t)
def h(t):
    return np.exp(-t**2 / 2) / np.sqrt(2 * np.pi)

# Generate t values in the interval [-3, 3]
t = np.linspace(-3, 3, 100)

# Compute the corresponding h(t) values
h_values = h(t)

# Plot the function
plt.plot(t, h_values)
plt.xlabel('t')
plt.ylabel('h(t)')
plt.title('Plot of h(t)')
plt.grid(True)
plt.show()


#(b) Implementing the gradient ascent algorithm with a learning rate of η = 0.05 and 
#starting value t(0) = -2, evaluating 15 iterations, and plotting the update values 
#against the iteration number:

import numpy as np
import matplotlib.pyplot as plt

# Define the function h(t) and its derivative
def h(t):
    return np.exp(-t**2 / 2) / np.sqrt(2 * np.pi)

def h_prime(t):
    return -t * h(t)

# Gradient Ascent Algorithm
def gradient_ascent(eta, t0, num_iterations):
    t_values = [t0]
    for i in range(num_iterations):
        t = t_values[-1]
        t_new = t + eta * h_prime(t)
        t_values.append(t_new)
    return t_values

# Set the learning rate, starting value, and number of iterations
eta = 0.05
t0 = -2
num_iterations = 15

# Perform gradient ascent and obtain the update values
t_values = gradient_ascent(eta, t0, num_iterations)

# Plot the update values against the iteration number
iterations = range(num_iterations + 1)
plt.plot(iterations, t_values, marker='o')
plt.xlabel('Iteration')
plt.ylabel('t')
plt.title('Gradient Ascent')
plt.grid(True)
plt.show()


#(c) Repeating part (b) with a learning rate of η = 0.2 and starting value t(0) = 1 to 
#observe the convergence behavior:

import numpy as np
import matplotlib.pyplot as plt

# Define the function h(t) and its derivative
def h(t):
    return np.exp(-t**2 / 2) / np.sqrt(2 * np.pi)

def h_prime(t):
    return -t * h(t)

# Gradient Ascent Algorithm
def gradient_ascent(eta, t0, num_iterations):
    t_values = [t0]
    for i in range(num_iterations):
        t = t_values[-1]
        t_new = t + eta * h_prime(t)
        t_values.append(t_new)
    return t_values

# Set the learning rate, starting value, and number of iterations
eta = 0.2
t0 = 1
num_iterations = 15

# Perform gradient ascent and obtain the update values
t_values = gradient_ascent(eta, t0, num_iterations)

# Plot the update values against the iteration number
iterations = range(num_iterations + 1)
plt.plot(iterations, t_values, marker='o')
plt.xlabel('Iteration')
plt.ylabel('t')
plt.title('Gradient Ascent')
plt.grid(True)
plt.show()
```
